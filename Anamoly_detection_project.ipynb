```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#label encode:

# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
data['xmlns']= label_encoder.fit_transform(data['xmlns'])

data['xmlns'].unique()
```

```python
data.head(5)

```

```python

# Encode labels in column 'species'.
data['Name']= label_encoder.fit_transform(data['Name'])

data['Name'].unique()
```

```python
data.head(15)
```

```python
# Encode labels in column 'species'.
data['Guid']= label_encoder.fit_transform(data['Guid'])

data['Guid'].unique()
```

```python
k_clusters = 2
labels = KMeans(k_clusters, init='k-means++', n_init=20, max_iter=500, random_state=4).fit_predict(data)
```

```python
df_clust = pd.DataFrame(data)
df_clust['Cluster'] = labels
df_clust.head()
```

```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sklearn as sk
import sklearn.linear_model as slm
import matplotlib.pyplot as plt
#from sklearn.model_selection import train_test_split ## Higher Version 0.19.1
from sklearn.model_selection import train_test_split
```

```python
X = df_clust.drop('Cluster', axis = 1)
y = data['Cluster']
LR = slm.LinearRegression()
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.25)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python

LR.fit(X_train, Y_train)

predict = LR.predict(X_test)
```
